```python
#i
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.DESCR)
```

```python
#ii
import pandas as pd
ds = pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
#iii
ds['MEDV'] = boston.target
```

```python
#iv
pd.DataFrame(ds.corr().round(3))
```

```python
#v
x=ds['RM']
y=ds['CRIM']
pd.DataFrame([x,y])
```

```python
#vi
from sklearn.model_selection import train_test_split
x=pd.DataFrame(x)
y=pd.DataFrame(y)
x_train_1,x_test_1,y_train_1,y_test_1=train_test_split(x,y,test_size=0.3)
```

```python
#vii
from sklearn.linear_model import LinearRegression
train_model=LinearRegression()
train_model.fit(x_train_1,y_train_1)
train_model.predict(x_test_1)
train_model.predict(y_test_1)
```

```python
#viii
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
print(np.sqrt(mean_squared_error(x_test_1,y_test_1)))
print(r2_score(x_test_1,y_test_1))
```

```python
#ix a.
from sklearn.linear_model import Ridge
Ridge(alpha=1.5).fit(x_train_1, y_train_1)
```

```python
#ix b.
ridge_reg2=Ridge(alpha=3.5)
ridge_reg2.fit(x_train_1, y_train_1)
y_pred = ridge_reg2.predict(x_test_1)
print(np.sqrt(mean_squared_error(y_test_1,y_pred)))
print(r2_score(y_test_1,y_pred))
```

```python
#x a.
from sklearn.linear_model import Lasso
Lasso(alpha=0.1).fit(x_train_1, y_train_1)
```

```python
#x b.
lasso_reg1=Lasso(alpha=1)
lasso_reg1.fit(x_train_1, y_train_1)
y_pred = lasso_reg1.predict(x_test_1)
print(np.sqrt(mean_squared_error(y_test_1,y_pred)))
print(r2_score(y_test_1,y_pred))
```
